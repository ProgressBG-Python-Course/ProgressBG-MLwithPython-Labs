```python
fig = plt.figure()
fig.set_dpi(100)

# fig.add_axes([1,1,1,1])
# fig.add_axes([1,1,1,2])

ax1 = fig.add_subplot(1, 2, 1, facecolor='#FF0000')
ax2 = fig.add_subplot(1, 2, 2, facecolor='#0F0')

ax1.set_facecolor('#99F')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
fig.dpi=100;
fig.tight_layout()
# print(axes)


labels_style = {
	"color":'yellow',
	"backgroundcolor":'#999'
}

axes[0].set_facecolor('#99F')
axes[1].set_facecolor('#9F9')

axes[0].set_title('Title 1', fontdict={
	# 'horizontalalignment':'left'
	'color':'#F90'
}, loc="right")

axes[0].set_xlabel('X', loc='right', fontdict=labels_style)
axes[0].set_ylabel('Y', rotation=0, loc='top', fontdict=labels_style)

axes[1].set_ylabel('Y')

plt.setp(axes[0].spines.values(), color="#F3FF00")

# plot line
X = np.linspace(1,10,5)
Y = np.sin(2*X)

axes[0].plot(X,Y)
plt.show()
print(X)



```

```python
# pyplot api:
x = np.linspace(0,10,100)

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))
plt.title('sin')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title('cos')
```

```python
n = 256
X = np.linspace(-np.pi,np.pi,n)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color=(0,1,0,1)) # equiv to: color='green', alpha=1.00

plt.show()
```
