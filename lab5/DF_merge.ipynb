```python
import pandas as pd
```

```python
# # Create DF:
#   did	dname
# 0	1	Ivan
# 1	2	Asen
# 2	3	Maria
# 3	4	Stoyan
# 4	5	Aleks
# 5	6	Svetlin
```

```python
names = [ "Ivan", "Asen", "Maria", "Stoyan", "Aleks", "Svetlin"] 

devs = pd.DataFrame(enumerate(names,1), columns=["did", "dname"]) 
devs
```

```python
# did	language
# 0	2	"C++"
# 1	3	"Python"
# 2	3	"R"
# 3	6	"Java"

langs = pd.DataFrame({
	'did': [2,3,3,6],
	'language': ["C++","Python","R","Java"]
})

langs
```

```python
inner = devs.merge(langs, how='inner', on='did')
inner
```

```python
full_outer = devs.merge(langs, how='outer', on='did')
full_outer
```

```python
left_outer = devs.merge(langs, how='left', on='did')
left_outer
```

```python
right_outer = devs.merge(langs, how='right', on='did')
right_outer
```
